```python
import os
import pandas as pd
```

```python
import re
def remove_specific_part(text):
    text = re.sub(r'#\d+ \d{4}-\d{1,2}-\d{1,2} (오전|오후) \d{1,2}:\d{2}', '', text)
    text = re.sub(r'(오전|오후) \d{1,2}:\d{2}','',text)
    text = re.sub(r'\d+일 / \d{2}:\d{2}', '', text)
    text = re.sub(r'Post 글쓴이 \w+','',text)
    text = re.sub(r'Post 글쓴이','',text)
    text = re.sub(r'\d{4}년', '', text)
    text = re.sub(r'\d{1,2}월', '', text)
    text = re.sub(r'\d{1,2}일', '', text)
    text = re.sub(r'\d{2}:\d{2}','',text)
    text = re.sub(r'\d{4}-\d{1,2}-\d{1,2}','',text)
    text = re.sub(r'#\d{5} \d{8}','',text)
    return text.strip()
```

# 16

```python
def processing(df):
    labels = df.iloc[:, 1]
    conversations = df.iloc[:, 2:]
    result = pd.DataFrame({
        'Label': labels.repeat(conversations.shape[1]).reset_index(drop=True),
        'Conversation': conversations.values.flatten()
    })
    result.dropna(inplace=True)
    result.reset_index(drop=True, inplace=True)
    result['Conversation'] = result['Conversation'].apply(remove_specific_part)
    return result
```

```python
# 데이터가 저장된 폴더 경로
folder_path_1 = 'data/16'

# 폴더 내의 모든 파일 목록 가져오기
file_list = os.listdir(folder_path_1)

# csv 파일만 선택하기
csv_files = [file for file in file_list if file.endswith('.csv')]

# 각 파일을 데이터프레임으로 불러오기
labeled_data_16 = pd.DataFrame(columns=['Label', 'Conversation'])

for file in csv_files:
    file_path = os.path.join(folder_path_1, file)
    tmp = pd.read_csv(file_path, header=None)
    tmp_data = processing(tmp)
    labeled_data_16 = pd.concat([labeled_data_16, tmp_data], ignore_index=True)
```
